```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    crime = pd.read_csv("crime_data.csv", index_col='Year')   #read in the data
    crime.plot(y='Murder')      #we will plot murder against year as line plot
    plt.ylabel('Amount of Murders')
    plt.title('Number of Murders by Year')
    plt.show()
    
    crime.plot(y='Aggravated Assault', style='r-')    #we will plot assault against year as line plot
    plt.ylabel('Incidents of Aggravated Assault (multiplied by 1e6)')
    plt.title('Incidents of Aggravated Assault by Year')   #label and title
    plt.show()
    
    crime['Burglary'].hist(grid=False, bins=10)      #create a histogram of burglaries
    plt.title('Number of Years With Certain Amounts of Burglaries')
    plt.ylabel('Number of Years')
    plt.xlabel('Amount of Burglaries (multiplied by 1e6)')    #label and title
    plt.show()
```

```python
prob1()
```

# Problem 2

```python
 def prob2():
    """
    Use a visualization for each of the following to identify if trends exists 
    between Forcible Rape and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Plot each visualization.
    Return a tuple of booleans. Each entry should identify whether that 
    element correlates or not.
    """
    crime = pd.read_csv("crime_data.csv", index_col='Year')   #read in the data
    pd.plotting.scatter_matrix(crime[['Forcible Rape','Violent']], figsize=(10,10))   #plot rape vs violent crime
    plt.suptitle('Forcible Rape vs Violent Crime')
    plt.tight_layout()
    plt.show()
    
    pd.plotting.scatter_matrix(crime[['Forcible Rape','Burglary']], figsize=(10,10))   #plot rape vs burglary
    plt.suptitle('Forcible Rape vs Burglary')
    plt.tight_layout()
    plt.show()
    
    pd.plotting.scatter_matrix(crime[['Forcible Rape','Aggravated Assault']], figsize=(10,10))  #plot rape vs assault
    plt.suptitle('Forcible Rape vs Aggravated Assault')
    plt.tight_layout()   #make sure the graphs are clean
    plt.show()
    
    return True, False, True
```

```python
prob2()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    
    crime = pd.read_csv("crime_data.csv", index_col='Year')   #read in the data
    
    crime.plot(kind="box", y=["Burglary","Violent","Vehicle Theft"])   #plot the 3 different crimes in whisker boxes
    plt.xlabel('Type of Crime')
    plt.ylabel('Number of Occurrences (multiplied by 1e6)')  #label and title the plot
    plt.title('Distribution Info of Various Crimes')
    plt.show()
    
    crime.plot(kind="hexbin", x="Vehicle Theft", y="Robbery", gridsize=15, sharex=False)   #plot theft and robbery in hexbin
    plt.title('Comparing Vehicle Theft and Robbery')
    plt.show()
```

```python
prob3()
```
